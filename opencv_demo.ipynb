```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('cameraman.jpg')
```

```python
plt.imshow(image)
```

```python
image.shape
```

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
print(gray_img.shape)
plt.imshow(gray_img, cmap='gray')
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(gray_img,-1,kernel)
```

```python
plt.subplot(121),plt.imshow(gray_img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(dst, cmap='gray'),plt.title('Result')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
laplacian = cv2.Laplacian(gray_img, cv2.CV_64F)

plt.subplot(121),plt.imshow(gray_img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(laplacian, cmap='gray'),plt.title('Result')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
sobelx = cv2.Sobel(gray_img, cv2.CV_64F, 1, 0, ksize=5)

plt.subplot(121),plt.imshow(gray_img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(sobelx, cmap='gray'),plt.title('Sobel X')
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
sobely = cv2.Sobel(gray_img, cv2.CV_64F, 0, 1, ksize=5)

plt.subplot(121),plt.imshow(gray_img, cmap='gray'),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(sobely, cmap='gray'),plt.title('Sobel Y')
plt.xticks([]), plt.yticks([])
plt.show()
```
